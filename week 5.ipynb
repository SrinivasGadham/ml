```python
import pandas as pd
df=pd.read_csv("employees.csv")
df
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
df['SALARY']
```

```python
df[['SALARY']]
```

```python
imp=SimpleImputer(strategy="mean")
imp=imp.fit(df[['EMPLOYEE_ID','MANAGER_ID']])
imp
```

```python
df['MANAGER_ID']=df['MANAGER_ID'].str.replace('-','0')
df['MANAGER_ID']
```

```python
imp=SimpleImputer(strategy="mean")
imp=imp.fit(df[['EMPLOYEE_ID','MANAGER_ID']])
imp
```

```python
df[['EMPLOYEE_ID','MANAGER_ID']]
```

```python
df[['EMPLOYEE_ID','MANAGER_ID']]=imp.transform(df[['EMPLOYEE_ID','MANAGER_ID']])
print(df)
```
